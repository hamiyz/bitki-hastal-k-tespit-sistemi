```python
import os

TRAIN_DIR = "/content/drive/MyDrive/plant_disease_project/Plant-Disease-Detection-and-Solution/data/output/train"
VAL_DIR = "/content/drive/MyDrive/plant_disease_project/Plant-Disease-Detection-and-Solution/data/output/val"

classes_train = sorted(list(filter(
    lambda x: os.path.isdir(os.path.join(TRAIN_DIR, x)),
    os.listdir(TRAIN_DIR)
)))

print("Sınıf sayısı:", len(classes_train))
print("İlk 10 sınıf:", classes_train[:10])
```

```python
batch_size = 128

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1/255)
val_datagen = ImageDataGenerator(rescale=1/255)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(200, 200),
    batch_size=batch_size,
    classes=classes_train,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(200, 200),
    batch_size=batch_size,
    classes=classes_train,
    class_mode='categorical',
    shuffle=False
)
```

```python
import tensorflow as tf

num_classes = len(classes_train)

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(200, 200, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

model.summary()
```

```python
import tensorflow as tf
from tensorflow import keras
import os

save_dir = "/content/drive/MyDrive/plant_disease_project/Plant-Disease-Detection-and-Solution/models"
os.makedirs(save_dir, exist_ok=True)

best_model_path = f"{save_dir}/best_model.h5"

model = tf.keras.models.load_model(best_model_path)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

checkpoint = keras.callbacks.ModelCheckpoint(
    best_model_path,
    monitor='val_loss',
    verbose=1,
    save_best_only=True,
    save_weights_only=False
)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    initial_epoch=6,
    epochs=12,
    callbacks=[checkpoint]
)
```

```python
import tensorflow as tf

base_model = tf.keras.applications.EfficientNetB0(
    input_shape=(224,224,3),
    include_top=False,
    weights='imagenet'
)

x = tf.keras.layers.Input(shape=(224,224,3))
y = tf.keras.applications.efficientnet.preprocess_input(x)  # 🔥 EKLE
y = base_model(y)
y = tf.keras.layers.GlobalAveragePooling2D()(y)
y = tf.keras.layers.Dense(num_classes, activation='softmax')(y)

model = tf.keras.Model(x, y)
```

```python
import tensorflow as tf

MODEL_PATH = "/content/efficientnet_no_leaf_best_20epoch.keras"

old_model = tf.keras.models.load_model(MODEL_PATH)

inputs = tf.keras.Input(shape=(224, 224, 3))
x = tf.keras.applications.efficientnet.preprocess_input(inputs)
outputs = old_model(x)

new_model = tf.keras.Model(inputs, outputs)

converter = tf.lite.TFLiteConverter.from_keras_model(new_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]

tflite_model = converter.convert()

with open("/content/model.tflite", "wb") as f:
    f.write(tflite_model)

print("model.tflite oluşturuldu ✅")
```
